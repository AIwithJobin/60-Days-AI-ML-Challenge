```python
d={'Name':["Anagha","Anju","Sakshi","Aruna","Vishnu"],
   'Subject' : ['sub1','sub2','sub3','sub4','sub5'],
   'Marks': [45,40,46,35,37]
   }
one=pd.DataFrame(d,index=[1,2,3,4,5])
```

```python
#pandas provide aggrigate agg() which allows you to apply one or more aggrigation
#aggrigation functions works on a group of data to produce a result
#Example : sum(),min(),max()
import pandas as pd
import numpy as np

d={'A':[1,4,7,3],
   'B' : [2,5,8,2],
   "C": [3,6,9,5],
   "D" :[6,1,2,4]
   }
df=pd.DataFrame(d,index=["R1","R2","R3","R4"])
```

```python
result
```
